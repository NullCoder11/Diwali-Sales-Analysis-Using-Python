```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding = 'unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
#data cleaning
df.info()
```

```python
#axis=vertical row
#inplace means implementing and saving the changes
#Here we're removing the data columns named Status and unnamed1

df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
#to check the total number of null values
pd.isnull(df).sum()
```

```python
df.shape
```

```python
#dropping the null values
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
#using describe() on specific columns
df[['Age','Orders','Amount']].describe()
```

_Exploratory Data Analysis

**GENDER

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

_From the above graph we can deduce that most of the buyers are females and even the purchasing amount is greater than men.

**AGE

```python
#age group division according to gender for age group

ax = sns.countplot(data = df, x = 'Age Group',hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

_From above we can deduce that most buyers are of the age group between 26-35 female

**STATE

```python
#total number of sales(orders) from top 10 states

sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,5)})   #(width,height)-for ease of reading
sns.barplot(data=sales_state,x ='State',y='Orders')
```

```python
#total number of sales(amount) from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

_From above we can deduce that most of the orders and total sales amount are from Uttar Pradesh, Maharashtra and Karnataka respectively.

**MARITAL STATUS

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(8,5)}) 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount',hue='Gender')
```

_From above we can deduce that most of the buyers are females and they have high purchasing power

**OCCUPATION

```python
ax = sns.countplot(data = df, x = 'Occupation')

sns.set(rc={'figure.figsize':(30,5)}) 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

_From above we can deduce that most of the purchase is from IT Sector and least are from Lawyer

**PRODUCT CATEGORY

```python
ax = sns.countplot(data = df, x = 'Product_Category')

sns.set(rc={'figure.figsize':(35,15)}) 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

_From above graphs we can see that most of the sold products are from Food, Clothing and Electronics Category

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

*CONCLUSION*

Married women of age group category 26-35 years from UP,Maharashtra and Karnataka working in IT,Healthcare and Aviation are more likely to buy products from food clothing and electronics category
